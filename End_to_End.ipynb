```python
#read the csv and start working with it keep genre keyword and drop all other object based columns while removing popularity and vote_count
movie = pd.read_csv('/content/drive/MyDrive/tmdb_5000_movies.csv')

display(movie)
movie.info()
```

```python
#drop the unneeded columns
print(movie.columns)
keep = movie[['budget', 'genres', 'original_language','revenue','runtime','vote_average']]

keep.columns
```

```python
#create a func that parses the json to a ocmma seperated list of genres
def json_parse(json_needing_parsed):
  s = json.loads(json_needing_parsed)
  return ','.join([r['name'] for r in s])

#parse the genres column

movie_genre = keep['genres'].apply(json_parse)
#get a set of genres for the movies 
list_genres = set(','.join(movie_genre).split(','))
```

```python
#add the genres to the dataframe 
for g in list(list_genres)[1:]:
  keep[g] = 0 
```

```python
keep.head()
```

```python
# create function that check if genre is in the list for each observation, if the genere is present we update it to 1
# iterate over each row and check to see if
def updateGenre(row):
  s = json.loads(row['genres']) ## converting this string to .json
  for r in s: # r is genre in this case
    row[r["name"]] = 1
  return row
```

```python
# use the apply function to iterate over each row
keep = keep.apply(updateGenre, axis=1)
keep.head()
```

```python
keep.drop(columns =['genres'], inplace=True)
keep.head()
```

```python
# get dummies to handle the language catagories
keep = pd.get_dummies(keep)
keep.head()
```

```python
# get rid of NaN
keep = keep.dropna()
keep.head()
```

```python
#labels for our data to be compared to
labels = keep['vote_average']
labels.head()
```

```python
labels.describe()
```

```python
#the data to feed our model
data = keep.drop(columns =['vote_average'])
data.head()
```

```python
#check the dtypes of the remaining data
keep.dtypes
```

```python
# split data

# to make this notebook's output identical at every run
RANDOM_STATE = 42
np.random.seed(RANDOM_STATE)
```

```python
#create the testing and training data
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(data, labels, test_size=0.2, \
                                                    random_state=RANDOM_STATE)
```

```python
#check the shape to make sure it is correct and symmetrical
print(X_train.shape, y_train.shape)
X_test.shape, y_test.shape
```

```python
from sklearn.preprocessing import MinMaxScaler
# use minmax to handle 1 and 0
scaler = MinMaxScaler()
scaled = scaler.fit_transform(X_train)
```

```python
# linear regression
from sklearn.linear_model import LinearRegression

lin_reg = LinearRegression()
lin_reg.fit(scaled, y_train)
```

```python
# check model
from sklearn.metrics import mean_squared_error

predictions = lin_reg.predict(scaled)
lin_mse = mean_squared_error(y_train, predictions)
lin_rmse = np.sqrt(lin_mse)
lin_rmse
```

```python
# do cross validation

from sklearn.model_selection import cross_val_score
from sklearn.linear_model import LinearRegression

lin_reg = LinearRegression()
scores = cross_val_score(lin_reg, scaled, y_train,
                         scoring="neg_mean_squared_error", cv=10)
rmse_scores = np.sqrt(-scores)
rmse_scores
```

```python
def display_scores(scores):
    print("Scores:", scores)
    print("Mean:", scores.mean())
    print("Standard deviation:", scores.std())

display_scores(rmse_scores)
```
